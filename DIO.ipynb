```python
import cv2
import numpy as np

# Lê a imagem
img = cv2.imread('OIP.jpeg')

# Converte para escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplica desfoque Gaussiano
suave = cv2.GaussianBlur(gray, (7, 7), 0)

# Limiariza a imagem
_, binarizada = cv2.threshold(suave, 160, 255, cv2.THRESH_BINARY_INV)

# Converte imagens em escala de cinza para BGR
suave_bgr = cv2.cvtColor(suave, cv2.COLOR_GRAY2BGR)
binarizada_bgr = cv2.cvtColor(binarizada, cv2.COLOR_GRAY2BGR)

# Cria uma imagem empilhada para visualização
resultado = np.vstack([
    np.hstack([suave_bgr, binarizada_bgr]),
    np.hstack([binarizada_bgr, cv2.bitwise_and(img, img, mask=binarizada)])
])

# Exibe o resultado
cv2.imshow("Binarização da imagem", resultado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
